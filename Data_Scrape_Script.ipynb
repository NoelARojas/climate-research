```python
def convert_day(num):
    if num < 10:
        return str(0) + str(num)
    else:
        return str(num)
```

```python
csv_file = open('usa_weather_data_2019_1.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Date', 'State', 'City', 'Zone ID', 'Min_Temperature', 'Mean_Temperature', 'Max_Temperature'])

for year in range(2019, 2020):
    for month in range(9, 13):
        day = calendar.monthrange(year, month)
        date_start = str(str(year)+convert_day(month)+convert_day(day[0]))
        date_end = str(str(year)+convert_day(month)+convert_day(day[1]))
        for airport in airports:
            source = 'https://api.weather.com/v1/location/' + airport[1] + ':9:US/observations/historical.json?apiKey=6532d6454b8aa370768e63d6ba5a832e&units=e&startDate=' + date_start + '&endDate=' + date_end
            data = requests.get(source)
            if data.status_code == 200:
                data = data.json()
                #print(source)

                count = 0 #len(data['observations'])
                total = 0
                temperature = []
                maximum = ''
                minimum = ''
                for obs in data['observations']:
                    temp = obs['temp']
                    try:
                        total += temp
                        temperature.append(temp)
                    except:
                        total =+ 0
                    count += 1
                    mean = str(total/count)
                    if len(temperature) > 0:
                        maximum = str(max(temperature))
                        minimum = str(min(temperature))
                        city = airport[0]
                        zone_id = airport[1]
                        state = airport[2]
                        date = str(str(year)+ "/" + convert_day(month))

                print(date, state, city)
                print("Mean= " + mean + "\nMax=" + maximum + "\nMin=" + minimum)
                csv_writer.writerow([date, state, city, zone_id, minimum, mean, maximum])

csv_file.close()
print("done!")
```
